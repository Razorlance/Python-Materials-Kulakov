# Week 3: Collaborative Workflows

## Theory

### Remote Repositories
A remote repository is a version of your project that is hosted on the internet or a network. It allows multiple developers to collaborate on the same project by pushing and pulling changes.

**Why use remote repositories?**
- To collaborate with other developers.
- To back up your project on a remote server.
- To facilitate continuous integration and deployment.


### Pull Requests
Pull requests are a feature provided by platforms like GitHub, GitLab, and Bitbucket. They allow developers to notify team members about changes they've pushed to a branch in a repository.

**Key benefits of pull requests:**
- Facilitate code review and discussion.
- Help maintain code quality and consistency.
- Enable collaboration before merging changes into the main branch.

## Key Points
- Remote repositories enable collaboration across different locations.
- Pull requests are a mechanism for merging changes after review.
- Understanding how to use remote repositories and pull requests is essential for team collaboration.


## Code and Examples
### Working with Remote Repositories


#### Adding a Remote Repository
```bash
# Add a remote repository
git remote add origin https://github.com/user/repo.git
```

*Example Usage:* Use git remote add to link your local repository to a remote repository hosted on platforms like GitHub.

#### Pulling Changes from Remote





```
# Pull changes from the remote repository and merge them into the current branch
git pull origin main

```



*Example Usage:* Use git pull to fetch and merge changes from the remote repository into your current branch.

#### Pushing Changes to Remote



```
# Push changes to the remote repository
git push origin main
```



*Example Usage:* Use git push to upload your local changes to the remote repository.

#### Creating a Pull Request
Push your branch to the remote repository.

1.   Open the repository on GitHub.

2.   Click on "Compare & pull request".

3.   Add a title and description for your pull request.

4.   Click on "Create pull request".

*Example Usage:* Use pull requests to propose changes to the codebase and discuss them with team members before merging.
